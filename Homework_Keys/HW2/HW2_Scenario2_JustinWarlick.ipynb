```python
import flopy
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mp
import os
%matplotlib inline
Lx = 100.
Ly = 100.
ztop = 0.
zbot = -50.
nlay = 1
nrow = 25
ncol = 25
dx = Lx/ncol
dy = Ly/nrow
dz = (ztop - zbot) / nlay
nper=1
steady=[True]



headobjL=flopy.utils.binaryfile.HeadFile('Warlick_HW2_S2L'+'.hds')
headL=headobjL.get_data(totim=1.0)
headobjR=flopy.utils.binaryfile.HeadFile('Warlick_HW2_S2R'+'.hds')
headR=headobjR.get_data(totim=1.0)
headobjB=flopy.utils.binaryfile.HeadFile('Warlick_HW2_S2B'+'.hds')
headB=headobjB.get_data(totim=1.0)

plt.figure(figsize=(10,5))
plt.grid()
x=np.arange(0,Lx,dx)
plt.plot(x,np.flipud(headL[0])[int(nrow/2)][:],label='Left Well Pumping')
plt.plot(x,np.flipud(headR[0])[int(nrow/2)][:],label='Right Well Pumping')
plt.plot(x,np.flipud(headB[0])[int(nrow/2)][:],label='Both Wells Pumping')
plt.title('Combined Head Transect across X-Domain',fontweight='bold',fontsize = 14)
plt.xlabel('X Distance (m)',fontsize=12)
plt.ylabel('Head (m)',fontsize=12)
plt.xticks(np.arange(0, 101, step=10))
plt.legend(loc='best')
plt.savefig('Transect_S2C.png')
plt.show()
```
